<h1>Matplotlib com OpenCV</h1>
<p>Matplotlib é uma biblioteca externa do Python que também pode ser utilizada com o OpenCV.</p>
<p>Essa biblioteca é muito utilizada para plotar 'figuras' que contêm gráficos ou imagens.</p>
<p>Algumas vantagens de sua utilização são:</p>
<ul>
    <li>Colocar várias imagens em uma mesma janela/'figura'</li>
    <li>Ver coordenadas</li>
    <li>Utilização de 'escala'</li>
    <li>Dar zoom</li>
</ul>
<p>Utilizaremos mais a função pyplot</p>
<p>Para isso podemos importar da seguinte maneira: <code>import matplotlib.pyplot as plt </code></p>
<p>Em OpenCV as imagens coloridas são lidas em BGR (blue/azul - green/verde - red/vermelho), por esse motivo para exibir a mesma imagem utilizando o maplotlib, é necessário converter a imagem - cv.cvtColor().</p>

<p>Os principais métodos que utilizaremos são:</p>
<ul>
    <li><b>plt.imshow()</b>: 'colocar' a imagem em uma janela.</li>
    <li><b>plt.show()</b>: mostrar a 'figura'/imagem(ns) na tela.</li>
    <li><b>plt.xticks()</b>: definir/retirar a 'régua'/esscala do eixo 'x'.</li>
    <li><b>plt.yticks()</b>: definir/retirar a 'régua'/esscala do eixo 'y'.</li>
    <li><b>plt.title()</b>: colocar título.</li>
    <li><b>plt.subplot()</b>: colocar várias imagens na mesma 'figura'.</li>
</ul>

```python
import cv2 as cv
import os 
import matplotlib.pyplot as plt
os.chdir("C:/Users/edub_/OneDrive/Área de Trabalho/PDI - IC/conteudo_p_teste")
```

```python
#exemplo utilizando matplotlib com OpenCV
img1 = cv.imread("lena.jpg")
img1 = cv.cvtColor(img1, cv.COLOR_BGR2RGB)
img2 = cv.imread("messi5.jpg")
img2 = cv.cvtColor(img2, cv.COLOR_BGR2RGB)
img3 = cv.imread("orange.jpg")
img3 = cv.cvtColor(img3, cv.COLOR_BGR2RGB)

titulos = ['lena', 'messi', 'orange']
imagens = [img1, img2, img3]

for i in range(len(imagens)):
    plt.subplot(1, 3, i+1)
    plt.imshow(imagens[i])
    plt.title(titulos[i])
    plt.xticks([]), plt.yticks([])
plt.show()
```

```python
#usando em escala de cinza
img1 = cv.imread("lena.jpg", 0)
img2 = cv.imread("messi5.jpg", 0)
img3 = cv.imread("orange.jpg", 0)

titulos = ['lena', 'messi', 'orange']
imagens = [img1, img2, img3]

for i in range(len(imagens)):
    plt.subplot(1, 3, i+1)
    plt.imshow(imagens[i], 'gray')
    plt.title(titulos[i])
    plt.xticks([0, 250, 500]), plt.yticks([]) #mostrando 'régua' no eixo x 
plt.show()
```
